```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, confusion_matrix
import matplotlib.pyplot as plt

columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
           'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']

# Function for loading data
def load_data(data):
    df = pd.read_csv(data, names = columns)
    return df

def clean_data(df):
    df = df.dropna()
    df['target'] = df['target'].apply(lambda x: 1 if x > 0 else 0)
    return df

def split_data(df, test_size = 0.2, random_state = 42): # Make sure 20% of data goes into testing, 80% goes into training
    X = df.drop('target', axis = 1) # Remove target from X (correct predictions)
    y = df['target'] # Only target
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size = test_size, random_state = random_state, stratify = y # Same proportion of non-disease to disease
    )
    return X_train, X_test, y_train, y_test

def scale_features(X_train, X_test):
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train) # Uses z-score normalization, mean = 0, standard deviation = 1
    X_test_scaled = scaler.transform(X_test) 
    return X_train_scaled, X_test_scaled, scaler

def train_model(X_train, y_train):
    model = LogisticRegression()  # Does Logistic Regression
    model.fit(X_train, y_train)
    return model

def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    
    
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred))

    return y_pred, y_proba



df = load_data(r'C:\Users\benja\OneDrive\Documents\GitHub\Heart-Disease-Classifier\data\processed.cleveland.data')
df = clean_data(df)
X_train, X_test, y_train, y_test = split_data(df)
X_train_scaled, X_test_scaled, scaler = scale_features(X_train, X_test)
model = train_model(X_train_scaled, y_train)
y_pred, y_proba = evaluate_model(model, X_test_scaled, y_test)
```
